```python
# solve with CVXPY
from scipy import sparse
import numpy as np
import numpy.linalg as LA
import scipy as sp
from sklearn.datasets import make_sparse_spd_matrix
import matplotlib.pyplot as plt
from cvxpy import *

np.random.seed(1)

# Problem data.
# n: Dimension of matrix.
n = 350 #250
alpha_ratio = 1
m = 1000  # Number of samples.
ratio = 0.9   # Fraction of zeros in S.

S_true = sparse.csc_matrix(make_sparse_spd_matrix(n, ratio))
R = sparse.linalg.inv(S_true).todense()
q_sample = np.real(sp.linalg.sqrtm(R)).dot(np.random.randn(n,m))
Q = np.cov(q_sample)
print('Q is positive definite? {}'.format(bool(LA.slogdet(Q)[0])))
mask = np.ones(Q.shape, dtype=bool)
np.fill_diagonal(mask, 0)
alpha_max = np.max(np.abs(Q)[mask])
alpha = alpha_ratio*alpha_max #0.001 for n=100, 0.01 for n=50
# np.linalg.norm(Q-Q.T)

## Solve with CVXPY.
S = Variable((n,n), PSD=True)
obj = -log_det(S) + trace(S*Q) + alpha*norm1(S)
prob = Problem(Minimize(obj))
prob.solve(solver='SCS',eps=1e-6,verbose=True)
cvxpy_obj = prob.value
cvxpy_S = S.value
```

## Check 1: OSQP/SCS actually succeed on coupled QP

```python
import numpy as np
from cvxpy import *

np.random.seed(1)
K = 8 # even 200 works # number of blocks
p = 50 # number of coupling constraints
nk = 300 # variable dimension of each subproblem QP
mk = 200 # constrain dimension of each subproblem QP
A_list = [np.random.randn(p, nk) for k in range(K)]
F_list = [np.random.randn(mk, nk) for k in range(K)]
q_list = [np.random.randn(nk) for k in range(K)]
x_list = [np.random.randn(nk) for k in range(K)]
g_list = [F_list[k].dot(x_list[k])+0.1 for k in range(K)]
A = np.hstack(A_list)
x = np.hstack(x_list)
b = A.dot(x)
P_list = [np.random.randn(nk,nk) for k in range(K)]
Q_list = [P_list[k].T.dot(P_list[k]) for k in range(K)]

var_name_lst = []
for k in range(K):
    var_name_lst.append('x'+str(k))
    
obj = 0
constr_lhs = 0
constr = []
for k in range(K):
    vars()[var_name_lst[k]] = Variable(nk)
    obj += quad_form(vars()[var_name_lst[k]],Q_list[k]) + q_list[k]*vars()[var_name_lst[k]]
    constr_lhs += A_list[k] * vars()[var_name_lst[k]]
    constr += [F_list[k]*vars()[var_name_lst[k]] <= g_list[k]]
constr += [constr_lhs == b]

prob1 = Problem(Minimize(obj), constr)
prob1.solve(solver='OSQP', eps_abs=1e-6, eps_rel=1e-8, verbose=True) 

prob2 = Problem(Minimize(obj), constr)
prob2.solve(solver='SCS', eps=1e-6, verbose=True) 
```

## Check 2: Slowness and inaccuracy of scipy.optimize.nnls on NNLS

```python
import numpy as np
from scipy import sparse
from scipy.optimize import nnls
import time
np.random.seed(1)
m, n = 10000, 8000
density = 0.001
X = sparse.random(m, n, density=density, data_rvs=np.random.randn)
y = np.random.randn(m)
```

```python
# Solve with scipy.optimize.nnls
t0 = time.time()
res = nnls(sparse.csr_matrix(X).todense(),y)
t1 = time.time()
print('run time = {}'.format(t1-t0))
print('constraint violation = {}'.format(np.min(res[0])))
print('objective value = {}'.format(np.linalg.norm(X.dot(res[0]) - y)))
```

```python
# Solve with OSQP
from cvxpy import *
beta = Variable(n)
obj = sum_squares(X*beta-y)
constr = [beta >= 0]
prob = Problem(Minimize(obj), constr)
prob.solve(solver='OSQP', verbose=True)
beta = beta.value
print('constraint violation = {}'.format(np.min(beta)))
print('objective value = {}'.format(np.linalg.norm(X.dot(beta)-y)))
```

```python
# Solve with SCS
from cvxpy import *
beta = Variable(n)
obj = sum_squares(X*beta-y)
constr = [beta >= 0]
prob = Problem(Minimize(obj), constr)
prob.solve(solver='SCS', eps=1e-6, verbose=True)
beta = beta.value
print('constraint violation = {}'.format(np.min(beta)))
print('objective value = {}'.format(np.linalg.norm(X.dot(beta)-y)))
```
